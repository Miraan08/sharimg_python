```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoint = [1,2,6,8]
ypoint = [3,8,1,10]
plt.plot(xpoint, ypoint)
plt.show()
```

```python
xpoint = [1,2,6,8]
ypoint = [3,8,1,10]
plt.plot(xpoint, ypoint,marker='o')
plt.grid()
plt.title("Graph")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = 1,4,9,16,25
y2 = [1,8,27,64,125]
plt.plot(x,y1,label="Squares")
plt.plot(x,y2, label="Cubes")
plt.legend()
plt.show()
```

```python
y = np.array([35,25,25,15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y,labels = mylabels)
plt.legend(title = "Four Fruits", loc = "upper left")
plt.show()
```

```python
y = np.array([35,25,25,15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
colors = ["Pink","Cyan","Skyblue","Yellow"]
plt.pie(y,labels = mylabels, colors = colors)
plt.legend(title = "Four Fruits", loc = "upper left")
plt.show()
```

```python
from numpy import random
x = random.choice([3,5,7,9], size =2) #[3,9];by default size is 1
print(x)
x = random.rand()                     # 1 float val bw 0 to 1
print(x)
x = random.rand(5)                    # 5 float val bw 0 to 1
print(x)
x = random.randint(2,100)             # 1 int val bw 2 to 100(both included)
print(x)
x = random.randint(100,size=5)        # 5 random int from 0 to 100
print(x)
x = random.randint(100,size=(3,5))    # 2D array of size 3X5
print(x)
```

```python
x = np.zeros(5)
print(x)
x = np.zeros(5, dtype=int)
print(x)
x = np.ones(5)
print(x)
x = np.ones(5, dtype=int)
print(x)
x = np.zeros((2,3),dtype=int)
print(x)
```

```python
#numpy.arange(start,stop,step,dtype)
x = np.arange(10,20,2)
print(x)
x = np.arange(5,dtype=float)
print(x)
#numpy.linspace(start,stop,step,dtype)
x =  np.linspace(10,20,5) # any 5 val in the range
print(x)
```
